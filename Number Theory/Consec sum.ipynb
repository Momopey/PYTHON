```python
def consec_sum(c,n):
    return int(c*n + n*(n+1)/2)

def pretty_print(c,n):
    text="(c:"+str(c)+" | n:"+str(n)+") "
    for i in range(int(n)):
        text+=str(int(c+i+1))+"+"*(i<n-1)
    text+=" = "+str(consec_sum(c,n))
    print(text)

def simplify_consec_range(c,n):
    if c<0:
        return (-1-c,n+c)
    return (c,n)

def find_consec_with_total(t):
    if t%2==1:
        c,n= simplify_consec_range((t-3)/2,2)
    else:
        c_half,n_half= find_consec_with_total(t/2)
        if n_half%2==1:
            c,n= simplify_consec_range(2*c_half+ (n_half+1)/2,n_half)
        if n_half%2==0:
            c,n= simplify_consec_range(c_half-n_half/2,2*n_half)    
    pretty_print(c,n)
    return(c,n)
```

```python
find_consec_with_total(123456)
```

```python
vals=[]
for x in range(1,2000):
    t=x
    print(t)
    (c,n)=find_consec_with_total(t)
    print(c,"-",n)
    print(consec_sum(c,n))
    vals.append((x,int(c),int(n)))
```
