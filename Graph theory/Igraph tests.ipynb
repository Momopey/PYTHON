```python
g=Graph()
n=10
g.add_vertices(n)
g.add_edges([(a,a+1) for a in range(n-1)]+[(n-1,0)])
vert_set=[list(range(len(g.vs)))]
extra_edge=[]
for i in range(5):
    face=math.floor(random.random()*len(vert_set))
    face_set=vert_set[face]
    if face_set!=[]:
        edge=(random.choice(face_set),random.choice(face_set))
        extra_edge.append(edge)
        g.add_edges([edge])
        print("vert set:"+str(vert_set))
        print("Edge removed:"+str(edge))
        vert_set.remove(face_set)
        left_set= face_set[min(edge):max(edge)]
        right_set=face_set[0:min(edge)]+face_set[max(edge):]
        if(left_set!=[]):
            vert_set.insert(face,left_set)
        if(right_set!=[]):
            vert_set.insert(face,right_set)

g.vs["label"]=[str(i) for i in range(n)]
print(g.vs["label"])
lay=["circle","drl","fr","kk","large"]
layout = g.layout(lay[0])
p=plot(g,layout=layout)
p.save("imag.png")
```
